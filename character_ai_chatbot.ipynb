Import OpenAI library.
load the .env file using dotenv.
Get the secret key from the .env file to create a connection with openai API.

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(api_key=os.getenv("OPEN_AI_SECRET_KEY"))
print("OpenAI client configured")
```

Get response from OpenAI using a message.
Function used - 
chat.completion.create()

You need to pass the message, model you want to use and the role that should be used.

```python
message = "What do you know about India?"
print(message)
```

```python
response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = [{"role": "user", "content": message}]
)

print(response.choices[0].message.content)
```
